funcion de transferencia en lazo abierto:

se identifica las variables

masa=m

factor de amortiguamient= $\zeta$

resorte=k

entrada(fuerza aplicada al sistema )=x(t)=F(t)

salida(desplazamiento de la masa)=y(t)

fuerza del resorte= f_k

fuerza del amortiguador f$_\zeta$

teniendo en cuenta que:

el resorte se opone al desplazamiento

el amortiguador se opone a la velocidad

la masa depende de la aceleracion decimos que:

la fuerza del resorte seria: f_k=-k.y(t), donde x(t):porsicion de la masa.

la fuerza del amortiguador sera: f$_\zeta$ =-k.y'(t),donde x'(t); velociadad de la masa.

la fuerza de la masa sera: f_m=k.y"(t), donde x"(t): aceleracion de la masa.

Aplicando la segunda ley de Newton para relacionar todas las fuerza:

$\sum F = m\,a$

se tiene que:

f(t)-$\zeta$y'(t)-ky(t)=my"(t)

dando orden

f(t)=my"(t)+$\zeta$y'(t)+ky(t)

aplicamos la trasformada de laplace asumiendo condiciones iniciales igual a cero

F(s)=ms$^2$Y(s)+$\zeta_s$Y(s)+kY(s)

factorizando

F(s)=Y(s)(ms$^2$+$\zeta_s$+k)

finalmente obtenemos la funcion de transferencia

$$H(s) = \dfrac{Y(s)}{F(s)} = \dfrac{1}{ms^2 + \xi s + k}$$

## Relacin

Vemos que la funcin de transferencia del circuito mecnico queda:

$$
H(s) = \frac{X(s)}{F(s)} = \frac{1}{ms^2 + \xi s + k}
$$

Y la del circuito elctrico:

$$
H(s) = \frac{V_o(s)}{V_i(s)} = \frac{1}{CLs^2 + \dfrac{L}{R} \cdot s + 1}
$$

Usamos la analoga fuerza-tensin:

#### Relacin entre los sistemas mecnico y elctrico

| Caracterstica              | Sistema Mecnico                  | Sistema Elctrico                |
|----------------------------|-----------------------------------|----------------------------------|
| Variable de entrada        | Fuerza $ f(t) $                 | Voltaje $ v(t) $               |
| Variable de salida         | Posicin $ x(t) $               | Corriente $ i(t) $             |
| Elemento inercial          | Masa $ m $                      | Inductor $ L $                 |
| Elemento disipativo        | Amortiguador $ b $              | Resistencia $ R $              |
| Elemento elstico          | Resorte $ k $                   | Capacitor $ 1/C $              |

sta relacin funciona ya que ambos sistemas son similares en estructura matemtica:

- En el sistema mecnico:
  - La masa opone inercia al movimiento.
  - El resorte almacena energa potencial.
  - El amortiguador disipa energa como calor/friccin.

- En el sistema elctrico:
  - El inductor almacena energa magntica (como la masa).
  - El capacitor almacena energa elctrica (como el resorte).
  - La resistencia disipa energa (como el amortiguador).

As, un circuito elctrico puede simular el comportamiento de un sistema mecnico .

$$
\underbrace{\frac{X(s)}{F(s)} = \frac{1}{ms^2 + bs + k}}_{\text{Sistema Mecnico}}
\quad \leftrightarrow \quad
\underbrace{\frac{V_o(s)}{V_i(s)} = \frac{1}{CLs^2 + \dfrac{L}{R} \cdot s + 1}}_{\text{Sistema electrico }}
$$

Esto permite estudiar sistemas mecnicos usando circuitos elctricos equivalentes, facilitando simulaciones, anlisis y diseo de controladores.

```python
%%writefile pages/1_AM_Modulacion.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq
from pydub import AudioSegment
import yt_dlp
import os
from io import BytesIO

st.set_page_config(page_title=" AM Modulation Explorer", layout="wide")
st.title(" Modulacin AM de tu cancin favorita")

url = st.text_input(" Ingresa el enlace de YouTube de tu cancin favorita:", "")

if url:
    ydl_opts = {
        "format": "bestaudio/best",
        "nocheckcertificate": True,
        "cookiesfrombrowser": ("chrome",),
        "quiet": True
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=False)
        audio_url = info["url"]





# Funciones auxiliares
def array_to_audiosegment(signal, sample_rate):
    signal_int16 = np.int16(signal / np.max(np.abs(signal)) * 32767)
    audio = AudioSegment(
        signal_int16.tobytes(),
        frame_rate=sample_rate,
        sample_width=2,
        channels=1
    )
    return audio

def export_audiosegment_to_bytes(audiosegment):
    buf = BytesIO()
    audiosegment.export(buf, format="wav")
    return buf.getvalue()

if url:
    try:
        with st.spinner(" Descargando y procesando audio... Esto puede tardar unos segundos"):
            # Descargar audio
            ydl_opts = {
                'format': 'bestaudio/best',
                'outtmpl': 'audio.%(ext)s',
                'postprocessors': [{
                    'key': 'FFmpegExtractAudio',
                    'preferredcodec': 'mp3',
                }],
                'quiet': True,
                'no_warnings': True,
            }
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                ydl.download([url])
            audio = AudioSegment.from_file("audio.mp3").set_channels(1).set_frame_rate(44100)
            fragment = audio[20000:25000]  # 5 segundos
            os.remove("audio.mp3")

            # Convertir a numpy
            samples = np.array(fragment.get_array_of_samples()).astype(np.float32) / 2**15
            Fs = fragment.frame_rate
            t = np.linspace(0, len(samples)/Fs, len(samples))

            # Seal mensaje
            m_t = samples

            # Parmetros portadora
            Ac = 1.0
            Fc = 10000  # Hz

            # Portadora
            c_t = Ac * np.cos(2 * np.pi * Fc * t)

            # Modulacin AM
            y_t = (1 + m_t) * c_t

            # Mezcla (seal modulada * portadora)
            mixed = y_t * c_t

            # FFT de la seal mezclada
            M = fft(mixed)
            f = fftfreq(len(mixed), 1/Fs)
            M_filtered = M.copy()
            M_filtered[np.abs(f) > 8000] = 0  # Filtro ideal
            demod = np.real(ifft(M_filtered))

        st.success(" Audio descargado y procesado correctamente!")

        # Conversin a audio
        audio_m = fragment
        audio_portadora = array_to_audiosegment(c_t, Fs)
        audio_modulada = array_to_audiosegment(y_t, Fs)
        audio_demodulada = array_to_audiosegment(demod, Fs)

        # Reproduccin
        st.subheader(" Escucha y compara las seales")
        st.audio(export_audiosegment_to_bytes(audio_m), format="audio/wav")
        st.caption(" **Seal Mensaje Original**")

        st.audio(export_audiosegment_to_bytes(audio_portadora), format="audio/wav")
        st.caption(" **Seal Portadora**")

        st.audio(export_audiosegment_to_bytes(audio_modulada), format="audio/wav")
        st.caption(" **Seal Modulada AM**")

        st.audio(export_audiosegment_to_bytes(audio_demodulada), format="audio/wav")
        st.caption(" **Seal Demodulada (filtrada)**")

        # Grficas en el tiempo
        st.subheader(" Grficas en el tiempo")
        fig, axs = plt.subplots(5, 1, figsize=(12, 12), sharex=True)
        axs[0].plot(t, m_t, color='navy'); axs[0].set_title("Seal Mensaje m(t)"); axs[0].grid()
        axs[1].plot(t, c_t, color='gray'); axs[1].set_title("Seal Portadora c(t)"); axs[1].grid()
        axs[2].plot(t, y_t, color='orange'); axs[2].set_title("Seal Modulada y(t)"); axs[2].grid()
        axs[3].plot(t, mixed, color='purple'); axs[3].set_title("Seal Mezclada y(t) * c(t)"); axs[3].grid()
        axs[4].plot(t, demod, color='green'); axs[4].set_title("Seal Demodulada (filtrada)"); axs[4].set_xlabel("Tiempo (s)"); axs[4].grid()
        st.pyplot(fig)

        # Espectros
        st.subheader(" Espectros de frecuencia")
        def plot_spectrum(signal, Fs, title):
            N = len(signal)
            spectrum = fft(signal)
            freqs = fftfreq(N, 1/Fs)
            fig, ax = plt.subplots(figsize=(10, 4))
            ax.plot(freqs[:N//2], np.abs(spectrum[:N//2]), color='darkred')
            ax.set_title(title)
            ax.set_xlabel("Frecuencia (Hz)")
            ax.set_ylabel("Magnitud")
            ax.grid()
            return fig

        st.pyplot(plot_spectrum(m_t, Fs, "Espectro de la Seal Mensaje"))
        st.pyplot(plot_spectrum(c_t, Fs, "Espectro de la Portadora"))
        st.pyplot(plot_spectrum(y_t, Fs, "Espectro de la Seal Modulada"))
        st.pyplot(plot_spectrum(mixed, Fs, "Espectro de la Seal Mezclada"))
        st.pyplot(plot_spectrum(demod, Fs, "Espectro de la Seal Demodulada"))

        # Comparacin final con seales desplazadas verticalmente
        st.subheader(" Comparacin de todas las seales (desplazadas para mejor visualizacin)")
        offsets = [0, 2, 4, 6, 8]  # Desplazamientos verticales
        signals = [m_t, c_t, y_t, demod, mixed]
        labels = ["Mensaje", "Portadora", "Modulada", "Demodulada", "Mezclada"]
        colors = ['navy', 'gray', 'orange', 'green', 'purple']

        fig2, ax2 = plt.subplots(figsize=(12, 6))
        for i, (sig, label, color) in enumerate(zip(signals, labels, colors)):
            ax2.plot(t, sig + offsets[i], label=label, color=color)
        ax2.set_xlabel("Tiempo (s)")
        ax2.set_title("Comparacin de seales (desplazadas verticalmente)")
        ax2.legend()
        ax2.grid()
        st.pyplot(fig2)

        st.markdown("""
        ---
         **Listo!** Has visualizado todo el proceso de modulacin y demodulacin AM, incluyendo seales y espectros.
        Prueba con otros enlaces y sigue explorando el mundo del procesamiento de seales!
        """)

    except Exception as e:
        st.error(f" Ocurri un error: {e}")
else:
    st.info(" Ingresa una URL para comenzar la exploracin musical con modulacin AM.")

```

**Anlisis Dinmico y Equivalencia entre Sistemas Fsicos**

Este dashboard presenta un modelo interactivo que simula el comportamiento dinmico de un sistema masa-resorte-amortiguador y su sistema elctrico equivalente (circuito LCR serie). La herramienta permite visualizar y comparar la respuesta de ambos sistemas ante variaciones en parmetros clave, como la frecuencia natural y el amortiguamiento.

**Caractersticas principales:**

1.  Modelo masa-resorte-amortiguador: Simula la dinmica de un sistema mecnico clsico de segundo orden.

2.  Sistema elctrico equivalente: Representacin analgica del sistema mecnico con elementos L, C y R.

3.  Control total de parmetros: El usuario puede elegir el tipo de sistema (subamortiguado, sobreamortiguado, crtico o inestable) y ajustar interactivamente la frecuencia natural () y el coeficiente de amortiguamiento ().

4.  Equivalencia visual: Presenta en paralelo los parmetros fsicos de ambos sistemas para facilitar su comparacin conceptual.

**Funcionalidades tcnicas:**

1.  Respuesta al escaln: Simulacin temporal de cmo responde el sistema ante una entrada tipo escaln unitario.

2.  Diagrama de Bode: Visualizacin de la respuesta en frecuencia (magnitud y fase) para ambos sistemas.

3.  Plano de polos y ceros: Muestra la localizacin de los polos en el plano complejo para evaluar estabilidad y comportamiento dinmico.

4.  Clculo de parmetros del sistema: Calcula los coeficientes del resorte, amortiguador, inductancia, capacitancia y resistencia a partir de las condiciones dadas.

**Control en tiempo real:**

El usuario puede:

1. Elegir el tipo de comportamiento dinmico (estable o inestable).

2. Ajustar el valor de amortiguamiento segn el tipo seleccionado.

3. Cambiar la frecuencia natural del sistema.

4. Observar cmo afectan estos cambios a los grficos y parmetros del sistema.

```python
%%writefile pages/2__Modulo_2.py

# Guarda el script en un archivo dentro del directorio pages/

import streamlit as st  # Librera para crear interfaces web interactivas
import numpy as np  # Librera para operaciones numricas y manejo de arrays
import matplotlib.pyplot as plt  # Librera para generar grficos
from control import TransferFunction, bode_plot, step_response  # Funciones para anlisis de sistemas de control
from control.matlab import step, impulse, pole  # Funciones adicionales de control (respuesta al impulso, polos, etc.)
import pandas as pd  # Librera para manipular y mostrar datos tabulares

# Configuracin de la pgina de Streamlit
st.set_page_config(page_title=" Sistema Mecnico y Elctrico", layout="wide")
st.title(" Sistema Mecnico y Elctrico")  # Ttulo principal de la aplicacin

# Sidebar: Seleccin de tipo de respuesta
st.sidebar.header("Configuracin del Sistema")

# Desplegable para seleccionar el tipo de respuesta del sistema
tipo_respuesta = st.sidebar.selectbox(
    "Tipo de Respuesta",
    ["Subamortiguada", "Sobreamortiguada", "Amortiguamiento Crtico", "Inestable"]
)

# Deslizadores para frecuencia natural y factor de amortiguamiento inicial
wn = st.sidebar.slider("Frecuencia Natural ()", 0.1, 20.0, 5.0)
zeta = st.sidebar.slider("Factor de Amortiguamiento ()", 0.0, 2.0, 0.5)

# Ajustar zeta segn el tipo de respuesta seleccionada
if tipo_respuesta == "Subamortiguada":
    zeta = st.sidebar.slider("", 0.0, 1.0, 0.3)
elif tipo_respuesta == "Sobreamortiguada":
    zeta = st.sidebar.slider("", 1.0, 2.0, 1.5)
elif tipo_respuesta == "Amortiguamiento Crtico":
    zeta = 1.0  # Valor fijo para amortiguamiento crtico
elif tipo_respuesta == "Inestable":
    zeta = st.sidebar.slider("", -1.0, 0.0, -0.1)

# Sistema mecnico: H(s) = 1 / (ms + bs + k)
m = 1.0  # Masa
k = wn**2  # Constante del resorte, relacionada con wn
b = 2 * zeta * wn * m  # Coeficiente de amortiguamiento

# Numerador y denominador de la funcin de transferencia mecnica
num_mec = [1]
den_mec = [m, b, k]
sys_mec = TransferFunction(num_mec, den_mec)  # Crear funcin de transferencia del sistema mecnico

# Sistema elctrico equivalente (LCR serie con salida en R)
L = m  # Inductancia anloga a la masa
C = 1 / k  # Capacitancia anloga al inverso de la constante del resorte
R = b  # Resistencia anloga al amortiguador

# Numerador y denominador del sistema elctrico equivalente
num_elec = [R*C, 0]
den_elec = [L*C, R*C, 1]
sys_elec = TransferFunction(num_elec, den_elec)  # Funcin de transferencia del sistema elctrico

# Mostrar parmetros del sistema en dos columnas
col1, col2 = st.columns(2)

# Parmetros del sistema mecnico
col1.subheader(" Sistema Mecnico")
col1.write(f"Masa (m): {m:.2f} kg")
col1.write(f"Amortiguador (b): {b:.2f} Ns/m")
col1.write(f"Resorte (k): {k:.2f} N/m")

# Parmetros del sistema elctrico
col2.subheader(" Sistema Elctrico")
col2.write(f"Inductor (L): {L:.2f} H")
col2.write(f"Capacitor (C): {C:.4f} F")
col2.write(f"Resistor (R): {R:.2f} ")

# Funcin para graficar el diagrama de Bode
def plot_bode(sys, title):
    w = np.logspace(-2, 2, 1000)  # Rango de frecuencias logartmico

    fig, ax = plt.subplots(2, 1, figsize=(10, 6))  # Crear figura con 2 subplots (magnitud y fase)
    try:
        bode_plot(sys, omega=w, dB=True, Hz=False, deg=True)  # Graficar diagrama de Bode
        plt.suptitle(title)  # Ttulo general
        st.pyplot(fig)  # Mostrar grfico en Streamlit
    except Exception as e:
        st.error(f" Error al graficar diagrama de Bode: {e}")  # Mostrar error si ocurre

# Funcin para graficar la respuesta al escaln
def plot_step(sys, title):
    try:
        t, y = step_response(sys)  # Obtener tiempo y respuesta
        plt.figure(figsize=(10, 4))
        plt.plot(t, y)  # Graficar la respuesta
        plt.grid()
        plt.xlabel('Tiempo (s)')
        plt.ylabel('Respuesta')
        plt.title(title)
        st.pyplot(plt)  # Mostrar grfico
    except Exception as e:
        st.error(f" Error al graficar respuesta al escaln: {e}")  # Mostrar error si falla

# Funcin para graficar polos y ceros (mapa PZ)
def plot_pzmap(sys, title):
    p = pole(sys)  # Obtener polos
    z = []  # No se calculan ceros explcitamente aqu
    plt.figure(figsize=(6, 6))
    plt.scatter(np.real(p), np.imag(p), marker='x', color='red', label='Polos')  # Graficar polos
    plt.axhline(0, color='black', lw=1)  # Lnea horizontal en imaginario = 0
    plt.axvline(0, color='black', lw=1)  # Lnea vertical en real = 0
    plt.grid(True)
    plt.legend()
    plt.title(title)
    st.pyplot(plt)  # Mostrar grfico

# Comentado: Clculo de parmetros temporales solo si es subamortiguado
# if tipo_respuesta == "Subamortiguada" and 0 < zeta < 1:
#     try:
#         wd = wn * np.sqrt(1 - zeta**2)  # Frecuencia amortiguada
#         tr = np.pi / wd  # Tiempo de levantamiento
#         tp = np.pi / wd  # Tiempo al pico
#         Mp = np.exp(-zeta * np.pi / np.sqrt(1 - zeta**2))  # Sobreimpulso
#         ts = 4 / (zeta * wn)  # Tiempo de establecimiento (criterio 2%)

#         st.markdown("##  Parmetros Temporales (Subamortiguado)")
#         params_df = pd.DataFrame({
#             "Parmetro": [
#                 "Tiempo de levantamiento (tr)",
#                 "Sobreimpulso mximo (Mp)",
#                 "Tiempo al pico (tp)",
#                 "Tiempo de establecimiento (ts)"
#             ],
#             "Valor": [
#                 f"{tr:.3f} s",
#                 f"{Mp*100:.2f}%",
#                 f"{tp:.3f} s",
#                 f"{ts:.3f} s"
#             ]
#         })
#         st.table(params_df)  # Mostrar tabla con parmetros
#     except Exception as e:
#         st.warning(" No se pudieron calcular los parmetros dinmicos.")  # Advertencia si hay error

# Ttulo para la seccin de anlisis
st.markdown("##  Anlisis del Sistema")
col1, col2 = st.columns(2)  # Crear dos columnas para mostrar los grficos

# Columna izquierda: Sistema Mecnico
with col1:
    st.markdown("###  Sistema Mecnico")
    plot_bode(sys_mec, "Bode - Sistema Mecnico")
    plot_pzmap(sys_mec, "Polos y Ceros - Sistema Mecnico")
    plot_step(sys_mec, "Respuesta al Escaln - Sistema Mecnico")

# Columna derecha: Sistema Elctrico
with col2:
    st.markdown("###  Sistema Elctrico")
    plot_bode(sys_elec, "Bode - Sistema Elctrico")
    plot_pzmap(sys_elec, "Polos y Ceros - Sistema Elctrico")
    plot_step(sys_elec, "Respuesta al Escaln - Sistema Elctrico")


```
