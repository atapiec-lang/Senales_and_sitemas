se tiene un microprocesador de 4 bits con entrada analoga entre -3.3 y 3.3 voltios describa las condiciones necesarias para que el microprocesador pueda digitalizar la señal X(t)=30cos(100PIt). presente una simulacion en python de dicho proceso para 3 ciclos de la señal x(t)

```python
import numpy as np
```

```python
# crear vector tiempo por rango y por número de muestras
K = 500 # número de muestras
ti = 0# tiempo incial referencia [s]
tf = 30 # tieimpo final  [s]
# vector de tiempo equiespacioado
tv = np.linspace(start=ti,stop=tf,num=K)
print(tv)
print(tv.shape)
# tiempo muestreo se define según el rango de trabajo en segundos y la cantidad de muestras
Ts = (tf - ti)/K
print(Ts)
```

```python
#parámetros de la senal
A = 30
F= 50
# creal la senal coseno en tiempo discreto
xv = A*np.cos(2*np.pi*F*tv) # creamos parabola en tiempo discreto-> **  pontencia
print(xv)
print(xv.shape)
```

```python
import matplotlib.pyplot as plt #librería para gráficas básicas en python

plt.plot(tv,xv,'g.') #vector en eje horizontal, vector eje vertical, color y marcador
plt.xlabel('tiempo [s]')
plt.ylabel('$x[t_k]$') # $ $ ecuación latex
plt.show() #mostrar gráfica
```

```python
y_min = -3.3 # valor min después de cero y pendiente
y_max = 3.3 # valor max después de cero y pendiente
#crear funcion en python
def my_ceropen(x,ymin = -5, ymax = 5): #en general se pueden definir valores por defecto
  '''
  Codigo base para simular proceso de cero y pendiente
  Se ingresa arreglo de numpy y los valores min  y max despues de cero y pendiente
  '''
  xmax = max(x) # x.max()
  xmin = min(x) # x.min()
  m = (ymax - ymin)/(xmax - xmin)
  c = ymin - m*xmin
  yv = m*x + c
  return yv
```

```python
yv = my_ceropen(xv,ymin=-3.3,ymax=3.3) #se llama la función definida en la celda anterior
```

```python
print(yv)
print(yv.shape)
```

```python
plt.subplot(121) # 1 = #filas, 2 = #columnas, 1 = # dibujo
plt.plot(tv,xv,'r.')
plt.xlabel('tiempo [s]')
plt.ylabel('$x[t_k]$')

plt.subplot(122)
plt.plot(tv,yv,'b.')
plt.xlabel('tiempo [s]')
plt.ylabel('$y[t_k]$')
plt.subplots_adjust(hspace=0.5,wspace=0.5) #ajustar espaciado entre gráficas
plt.show()
```
